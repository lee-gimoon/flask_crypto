# 람다 함수에 대하여...

람다 함수는 파이썬에서 익명의 간단한 함수를 생성하기 위한 방법입니다. `lambda` 키워드를 사용하여 정의되며, 주로 짧은 수명을 가진 함수에 사용됩니다.<br>
즉, 함수의 이름을 정의할 필요 없이 일회용 함수를 빠르게 생성할 수 있습니다.

### 람다 함수의 기본 구조:
```python
lambda arguments: expression
```

- `lambda`: 람다 함수를 정의하는 키워드입니다.
- `arguments`: 함수의 입력 인자입니다. 여러 개의 인자를 사용할 수 있습니다.
- `expression`: 인자를 사용한 표현식입니다. 이 표현식의 결과가 람다 함수의 반환값이 됩니다.

### 람다 함수의 예:

1.**기본적인 람다 함수**:
```python
f = lambda x: x + 10
print(f(5))  # 출력: 15
```
<br>

2.**여러 개의 인자를 가진 람다 함수**:
```python
g = lambda x, y: x + y
print(g(2, 3))  # 출력: 5
```
<br>

3.**리스트 정렬에 사용하는 람다 함수**:
```python
points = [(1, 2), (3, 4), (5, 1)]
sorted_points = sorted(points, key=lambda p: p[1])
print(sorted_points)  # 출력: [(1, 2), (5, 1), (3, 4)]
```

### 람다 함수의 주요 특징:
- 이름 없이 정의됩니다 (익명의 함수).
- 간단한 로직을 갖는 함수에 주로 사용됩니다.
- 주로 일회성의 함수를 만들 때 사용되며, 특히 `map()`, `filter()`, `sorted()`와 같은 내장 함수의 인자로 전달될 때 유용합니다.

그러나, 람다 함수는 간결성을 위해 사용되기 때문에 복잡한 로직이나 여러 표현식을 포함하는 함수를 작성하기에는 적합하지 않습니다. 이러한 경우에는 표준 함수 정의를 사용하는 것이 더 적합합니다.
