# asyncio.run() 함수를 사용하여 코루틴을 실행할 때의 작동 원리와 실행 흐름에 대한 간략한 개요입니다.

1. **이벤트 루프 생성**
   - `asyncio.run()`이 호출될 때, 현재 스레드에 대한 새로운 이벤트 루프가 생성됩니다.
   
2. **코루틴 실행**
   - 주어진 코루틴이 이벤트 루프에서 실행되기 시작합니다. 

3. **await 사용**
   - 코루틴 내부에서 `await` 키워드를 만나게 되면, 현재 코루틴은 일시 중지되고 이벤트 루프는 다른 작업(다른 코루틴 또는 태스크)을 실행할 수 있습니다.

4. **코루틴 재개**
   - `await`로 기다리고 있던 작업이 완료되면, 중지된 코루틴이 재개되고 실행을 계속합니다.

5. **이벤트 루프 종료**
   - 주어진 코루틴과 모든 관련 작업이 완료되면, 이벤트 루프는 종료됩니다.
   
6. **asyncio.run() 종료**
   - `asyncio.run()` 함수는 이벤트 루프가 종료되고 나면 반환됩니다.
